# **Question 1**

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt

image = cv2.imread('images.jpg', cv2.IMREAD_GRAYSCALE)

levels = int(input("How many intensity levels: "))

level = 256 / levels
new_image = np.floor_divide(image, level) * level
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(new_image, cmap='gray')
plt.axis('off')

plt.show()
```

# **Question 2**

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')
block_sizes = [3, 10, 20]
average_images = []

for block_size in block_sizes:
    average_image = cv2.blur(image, (block_size, block_size))
    average_images.append(average_image)

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

for i, block_size in enumerate(block_sizes):
    plt.subplot(2, 3, i + 2)
    plt.imshow(average_images[i], cmap='gray')
    plt.title(f'{block_size}x{block_size} Average')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Question 3**

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')
rows, cols, colors = image.shape
M_45 = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)
rotated_45 = cv2.warpAffine(image, M_45, (cols, rows))

M_90 = cv2.getRotationMatrix2D((cols / 2, rows / 2), 90, 1)
rotated_90 = cv2.warpAffine(image, M_90, (cols, rows))

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(rotated_45, cmap='gray')
plt.title('Rotated 45 Degrees')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(rotated_90, cmap='gray')
plt.title('Rotated 90 Degrees')
plt.axis('off')

plt.tight_layout()
plt.show()
```
